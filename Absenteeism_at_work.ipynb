# Data Preprocessing

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Absenteeism time in hours'] = scaler.fit_transform(df[['Absenteeism time in hours']])
```

## Encode Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder
obj_cols = df.select_dtypes(object).columns
for col in obj_cols:
  encoder = LabelEncoder()
  df[col] = encoder.fit_transform(df[col])

df.dtypes
```

## Handling Missing Values

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

# Split the Data

```python
x = df.drop(columns = ['Absenteeism time in hours'])
y = df['Absenteeism time in hours']
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size = 0.8)
```

# Model Selection

```python
from sklearn.metrics import mean_squared_error

def eval_model(model, xtrain, ytrain, xtest, ytest):
  model.fit(xtrain, ytrain)
  trainpred = model.predict(xtrain)
  testpred = model.predict(xtest)

  return mean_squared_error(ytrain, trainpred), mean_squared_error(ytest, testpred)
```

```python
result = pd.DataFrame(columns = ['Model Number', 'Name', 'Training Error', 'Testing Error'])
result.head()
```

```python
from sklearn.svm import SVR
model = SVR()
row = []
row.extend([1, 'SVR'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
row = []
row.extend([2, 'Random Forest'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor()
row = []
row.extend([3, 'GBM'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
from sklearn.linear_model import BayesianRidge
model = BayesianRidge()
row = []
row.extend([4, 'Bayesian Regression'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
from sklearn.linear_model import ElasticNet
model = ElasticNet()
row = []
row.extend([5, 'Elastic Net'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
from sklearn.kernel_ridge import KernelRidge
model = KernelRidge()
row = []
row.extend([6, 'Kernel Ridge'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
from sklearn.ensemble import AdaBoostRegressor
model = AdaBoostRegressor()
row = []
row.extend([7, 'AdaBoost'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
from sklearn.linear_model import QuantileRegressor
model = QuantileRegressor(solver='highs') # Specify the 'highs' solver
row = []
row.extend([8, 'Quantile Regression'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
row = []
row.extend([9, 'Decision Tree'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor()
row = []
row.extend([10, 'KNN Regression'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
row = []
row.extend([11, 'Extra Trees'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head(11)
```

```python
import numpy as np

models = result['Name']
training_accuracies = np.round(result['Training Error'],7)
testing_accuracies = np.round(result['Testing Error'],7)

x = np.arange(len(models))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 8))
rects1 = ax.bar(x - width/2, training_accuracies, width, label='Training Error')
rects2 = ax.bar(x + width/2, testing_accuracies, width, label='Testing Error')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Model')
ax.set_ylabel('Accuracy')
ax.set_title('Training and Testing Error of Various ML Models')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Function to add labels on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(7, 7),  # 7 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
